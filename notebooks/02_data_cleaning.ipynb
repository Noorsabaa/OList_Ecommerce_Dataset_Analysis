```python
def missing_data_summary(dataset):
    missing = dataset.isnull().sum()
    missing_percentage = (missing/len(dataset))*100
    missing_data = pd.DataFrame({'Missing Values': missing, 'Percentage': missing_percentage})
    return missing_data


for name, dataset in Datasets.items():
    print(f"Missing Data Summary for {name}:")
    display(missing_data_summary(dataset))
    print("\n")
```

```python
def duplicate_data_summary(dataset):
    duplicates = dataset.duplicated().sum()
    duplicate_percentage = (duplicates / len(dataset)) * 100

    return pd.DataFrame({
        "Metric": ["Duplicate Rows", "Duplicate Percentage"],
        "Value": [duplicates, round(duplicate_percentage, 2)]
    })

for name, dataset in Datasets.items():
    print(name)
    display(duplicate_data_summary(dataset))
    print("\n")
```

### Deduplicating the Geolocation Table

The geolocation table contains multiple latitude and longitude coordinates for the same ZIP code prefix. This is expected because a ZIP code prefix can cover many different addresses within the same area.

If we merge this table directly with customers or sellers using the ZIP code prefix, a single customer could match multiple geolocation rows. This causes a **fan-out problem**, where one original row expands into several rows after the merge.

Fan-out can inflate row counts and produce incorrect business metrics, such as overstated sales totals or duplicated customer records.

To prevent this, we keep only one representative coordinate for each ZIP code prefix using `drop_duplicates(subset='geolocation_zip_code_prefix')`. This ensures a one-to-one relationship between ZIP code prefixes and geolocation records, making future joins safe and preventing duplicate rows from being introduced.
